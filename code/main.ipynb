# 读取目标表和替换表

```python
path_ori = r'../src/最后匹配结果.xlsx'
ori = pd.read_excel(path_ori)
print(ori.columns)
ori = ori[~(ori.策略=='删除')]
ori = ori.drop(['情况','策略'], axis=1)
print(len(ori))
ori_keep = ori[ori.材料编码_y==ori.材料编码_y]
ori_need_rep = ori[~(ori.材料编码_y==ori.材料编码_y)]
```

```python
ori_need_rep = ori_need_rep.drop(['材料编码_y','材料名称', '材料代号', '主材编码', '主材大类', '主材中类', '主材小类'], axis=1)
```

```python
path_mm = r'../src/物料替换及主材标签_干变_配变_20210818.xlsx'
mm = pd.read_excel(path_mm)
mm.rename(columns={'材料编码':'材料编码_x','替换物料编码':'材料编码_y'}, inplace=True)
print(mm.columns)
```

```python
ori_need_rep = ori_need_rep.merge(mm, on='材料编码_x', how='left')
ori_need_rep.columns
```

```python
ori = pd.concat([ori_keep,ori_need_rep], sort=False)
ori.columns
```

```python
# path_output=r'../output/最后匹配结果_01_物料替换.xlsx'
# ori.to_excel(path_output, index=False)
```

以上步骤完成了对辅材和配件的替换。

对每一个TZB的图号添加对应容量的风机、外壳、温控。

```python
df_tzb = ori[ori.地点=='TZB']
df_tzb['容量'] = df_tzb.图号名称.apply(lambda x: re.findall(r'[1-9]+\.?[0-9]*', x)[1]).astype(int)
```

```python
df_tzb_append = df_tzb[['图号','容量']].drop_duplicates()
df_tzb_append
```

# 读取配件表

```python
path_file=r'../src/风机_温控_外壳_按容量分.xlsx'
df_rep = pd.read_excel(path_file)
df_rep.columns
```

```python
df_tzb_append = df_tzb_append.merge(df_rep,on='容量', how='left')
df_tzb_append = df_tzb_append.drop(['容量'], axis=1)
df_tzb_append.rename(columns={'材料编码':'材料编码_x'}, inplace=True)
df_tzb_append['材料编码_y'] = df_tzb_append.材料编码_x
df_tzb_append
```

```python
df_tzb_repd = df_tzb_append.merge(df_tzb[['图号编码','地点','图号名称','图号']].drop_duplicates(), on = '图号', how='left')
```

```python
result = pd.concat([df_tzb_repd, ori], sort=False)
```
